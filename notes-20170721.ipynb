### Count words in a text

```python
text = open("frost.txt").read()
```

```python
text
```

```python
print(text)
```

```python
words = text.split()
```

```python
words
```

```python
chars = list(text.lower())
```

```python
chars
```

```python
count_words = Counter(words)
```

```python
count_words["the"]
```

```python
count_words.most_common()
```

```python
count_words.most_common(10)
```

```python
count_chars = Counter(chars)
```

```python
count_chars.most_common()
```

### Revisiting dog colors

```python
import csv
```

```python
dogs = list(csv.reader(open("dogs-of-nyc.csv")))
```

```python
dogs[0]
```

```python
dogs[12000][4]
```

```python
colors = [row[4] for row in dogs[1:]]
```

```python
color_count = Counter(colors)
```

```python
color_count.most_common()
```

what I want to end up with:

    BLACK,23578
    WHITE,18621
    BROWN,9181
    ...

```python
print("color,count")
for color, count in color_count.most_common(10):
    print(color+","+str(count))
```
